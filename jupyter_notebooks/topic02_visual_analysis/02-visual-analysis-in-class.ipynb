# Взаимодействия
## Количественный — количественный

```python
feat = [f for f in df.columns if 'charge' in f]
df[feat].hist()
```

```python
sns.pairplot(df[feat])
```

```python
clr = df['Churn'].map({False: 'blue', True: 'orange'});
plt.scatter(df['Total eve charge'], df['Total intl charge'],
           color = clr, label = 'dsfd');
plt.xlabel('Вечерние начисления');
plt.ylabel('Междунар. начисления');
plt.legend();
```

```python
clr = df['Churn'].map({False: 'blue', True: 'orange'});
plt.scatter(df[df['Churn']]['Total eve charge']], df[df['Churn']]['Total intl charge'],
           color = clr, label = 'dsfd');
plt.xlabel('Вечерние начисления');
plt.ylabel('Междунар. начисления');
plt.title('Распределение ушедших и лояльных');
plt.legend();
```
